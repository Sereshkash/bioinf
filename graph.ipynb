```python
f = open('res_comp1.txt')
s = f.readlines()
f.close()
x1 = [0]
y1 = [1000]
for i in range(len(s)):
    x1.append(float(s[i].split()[2][:-1]))
    y1.append(int(s[i].split()[5][:-1]))
x1 = np.array(x1)
y1 = np.array(y1)

f2 = open('res_comp_with_repeat.txt')
s2 = f2.readlines()
f2.close()
x2 = [0]
y2 = [1000]
for i in range(len(s2)):
    x2.append(float(s2[i].split()[2][:-1]))
    y2.append(int(s2[i].split()[5][:-1]))
x2 = np.array(x2)
y2 = np.array(y2)

f2 = open('res_comp_without_repeat.txt')
s2 = f2.readlines()
f2.close()
x3 = [0]
y3 = [1000]
for i in range(len(s2)):
    x3.append(float(s2[i].split()[2][:-1]))
    y3.append(int(s2[i].split()[5][:-1]))
x3 = np.array(x3)
y3 = np.array(y3)


f2 = open('res_comp_with_increase_6000.txt')
s2 = f2.readlines()
f2.close()
x4 = [0]
y4 = [1000]
for i in range(len(s2)):
    x4.append(float(s2[i].split()[2][:-1]))
    y4.append(int(s2[i].split()[5][:-1]))
x4 = np.array(x4)
y4 = np.array(y4)

plt.plot(x3, y3, c = 'g', label = 'without repeat, N = 400')
plt.plot(x2, y2, c = 'b', label = 'with 3 random subseq in 6 time, N = 400')
plt.plot(x4, y4, c = 'r', label = 'with repeat and increase, N = 60000')
plt.plot(x1, y1, label = 'without increase, N = 400')
plt.legend()
plt.xlabel('x')
plt.ylabel('N')
plt.show()
```

```python
x_l = 2.0
number_set_of_subseq = 8000
f = open('data_res_comp_with_repeat_6_x2.txt', 'a')
f.write('x = {}:\n'.format(x_l))
f.close()

max_N = 0
max_N_index = 0
set_N = ''
for j in range(1, number_set_of_subseq):

    name_folder = 'iter/' + 'it' + str(j) + '/'
    if os.path.isdir(name_folder):
        name_set_res = name_folder + 'name_set_res.txt'
        f = open(name_set_res, 'r')
        s = f.readlines()
        f.close()
        N_j = len(s) // 2
        set_N += str(N_j) + ' '
        if N_j > max_N:
            max_N = N_j
            max_N_index = j

f = open('data_res_comp_with_repeat_6_x2.txt', 'a')
f.write('{}\n'.format(set_N))
f.close()


s = 'x = {}, N = {}, max_N_index = {}\n'.format(x_l, max_N, max_N_index)
f = open('res_comp_with_repeat_6_x2', 'a')
f.write(s)
f.close()

```

```python
name_folder = 'increase/'
f = open(name_folder + 'first_subseq.txt', 'r')
first_subseq = f.readlines()
f.close()

f = open(name_folder + 'second_subseq4.txt', 'r')
second_subseq = f.readlines()
f.close()

f = open(name_folder + 'second_subseq5.txt', 'r')
second_subseq1 = f.readlines()
f.close()

f = open(name_folder + 'sw_subseq.txt', 'r')
second_subseq2 = f.readlines()
f.close()

first_subseq_len = np.array([len(first_subseq[i]) for i in range(1, len(first_subseq), 2)])
second_subseq_len = np.array([len(second_subseq[i]) for i in range(1, len(second_subseq), 2)])
second_subseq_len1 = np.array([len(second_subseq1[i]) for i in range(1, len(second_subseq1), 2)])
second_subseq_len2 = np.array([len(second_subseq2[i]) for i in range(1, len(second_subseq2), 2)])

plt.xlabel('len subseq')
plt.ylabel('number of subseq')
plt.hist(first_subseq_len, alpha = 0.8, label = 'before increase')
plt.hist(second_subseq_len, alpha = 0.8, bins = 30, label = 'after increase 200 - 200')
plt.hist(second_subseq_len1, alpha = 0.8, bins = 30, label = 'after increase_msa 200 - 200')
plt.hist(second_subseq_len2, alpha = 0.8, bins = 30, label = 'after auto increase_msa')
plt.title('Number before = {}, after increase msa auto = {}'.format(len(first_subseq_len), len(second_subseq_len2)))
plt.legend()
plt.show()
```
