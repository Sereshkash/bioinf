```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
f = open('res_comp1.txt')
s = f.readlines()
f.close()
x1 = [0]
y1 = [1000]
for i in range(len(s)):
    x1.append(float(s[i].split()[2][:-1]))
    y1.append(int(s[i].split()[5][:-1]))
x1 = np.array(x1)
y1 = np.array(y1)

f2 = open('res_comp_with_repeat.txt')
s2 = f2.readlines()
f2.close()
x2 = [0]
y2 = [1000]
for i in range(len(s2)):
    x2.append(float(s2[i].split()[2][:-1]))
    y2.append(int(s2[i].split()[5][:-1]))
x2 = np.array(x2)
y2 = np.array(y2)
plt.plot(x2, y2, c = 'b', label = 'new with 3 random subseq in 6 time')
plt.plot(x1, y1, c = 'r', label = 'old')
plt.legend()
plt.xlabel('x')
plt.ylabel('N')
plt.show()
```

```python
f = open('data_res_comp.txt', 'r')
s = f.readlines()
f.close()
x_name = list(map(lambda d: d[:-2], s[::2]))
x_data = list(map(lambda f: f.split(), list(map(lambda d: d[:-2], s[1::2]))))
x_data_np = np.array(x_data).astype('int')

f = open('data_res_comp_with_repeat.txt', 'r')
s = f.readlines()
f.close()
x_name2 = list(map(lambda d: d[:-2], s[::2]))
x_data2 = list(map(lambda f: f.split(), list(map(lambda d: d[:-2], s[1::2]))))
x_data_np_2 = np.array(x_data2).astype('int')

plt.hist(x_data_np[12], bins = 20, rwidth = 0.5, alpha = 0.7, label = 'old')  
plt.hist(x_data_np_2[0], bins = 20, rwidth = 0.5, alpha = 0.7, label = 'new with 3 random subseq in 6 time')
plt.legend()
plt.xlabel(x_name2[0])
plt.show()
```
